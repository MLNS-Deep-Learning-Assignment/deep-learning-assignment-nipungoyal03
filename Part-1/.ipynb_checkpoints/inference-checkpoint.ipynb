```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
```

```python
data_files = ['data0.npy', 'data1.npy', 'data2.npy']
label_files = ['lab0.npy', 'lab1.npy', 'lab2.npy']
```

```python
datasets = []
for data_file, label_file in zip(data_files, label_files):
    data = np.load(data_file)
    labels = np.load(label_file)
    X = torch.FloatTensor(data).unsqueeze(1) / 255.0
    y = torch.FloatTensor(labels)
    datasets.append((data, labels, X, y))
```

```python
class SimpleCNN(nn.Module):
    def __init__(self, height, width):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool2d(2)
        h1, w1 = height // 2, width // 2
        
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(2)
        h2, w2 = h1 // 2, w1 // 2
        
        self.flat_size = 32 * h2 * w2
        self.fc = nn.Linear(self.flat_size, 1)
        
    def forward(self, x):
        x = self.pool1(torch.relu(self.conv1(x)))
        x = self.pool2(torch.relu(self.conv2(x)))
        x = x.view(-1, self.flat_size)
        return self.fc(x)
```

```python
checkpoint = torch.load('cnn_model.pth')
model = SimpleCNN(checkpoint['height'], checkpoint['width'])
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()
```

```python

def evaluate_dataset(X, y, data, name="Dataset"):
    with torch.no_grad():
        predictions = model(X).squeeze().numpy()
    actual = y.numpy()
    
    mse = mean_squared_error(actual, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(actual, predictions)
    
    tolerances = [0.5, 1.0, 2.0]
    accuracies = {f"Accuracy (±{tol})": calculate_accuracy(actual, predictions, tol) 
                  for tol in tolerances}
    
    errors = np.abs(predictions - actual)
    mean_error = np.mean(errors)
    median_error = np.median(errors)
    std_error = np.std(errors)
    
    print(f"\n{name}")
    print(f"Mean Squared Error: {mse:.4f}")
    print(f"Root Mean Squared Error: {rmse:.4f}")
    print(f"R² Score: {r2:.4f}")
    
    print(f"\n{name} Accuracy Metrics")
    for name, acc in accuracies.items():
        print(f"{name}: {acc:.2f}%")
    
    print(f"\n{name} Error Statistics")
    print(f"Mean Absolute Error: {mean_error:.4f}")
    print(f"Median Absolute Error: {median_error:.4f}")
    print(f"Standard Deviation of Error: {std_error:.4f}")
    
    return predictions, actual, errors

def calculate_accuracy(y_true, y_pred, tolerance):
    wt = np.abs(y_true - y_pred) <= tolerance
    return np.mean(wt) * 100
```

```python
all_predictions = []
all_actuals = []
all_errors = []

for i, (data, labels, X, y) in enumerate(datasets):
    predictions, actual, errors = evaluate_dataset(X, y, data, f"Dataset {i}")
    all_predictions.extend(predictions)
    all_actuals.extend(actual)
    all_errors.extend(errors)
```

```python


all_predictions = np.array(all_predictions)
all_actuals = np.array(all_actuals)
all_errors = np.array(all_errors)

plt.figure(figsize=(15, 15))

plt.figure()
plt.scatter(all_actuals, all_predictions, alpha=0.5)
plt.plot([min(all_actuals), max(all_actuals)], 
         [min(all_actuals), max(all_actuals)], 'r--', 
         label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Combined Datasets')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
num_examples = 3
for dataset_idx, (data, labels, _, _) in enumerate(datasets):
    indices = np.random.choice(len(data), num_examples, replace=False)
    start_idx = dataset_idx * len(data)
    
    for i, idx in enumerate(indices):
        plt.subplot(3, num_examples, dataset_idx * num_examples + i + 1)
        plt.imshow(data[idx])
        pred = all_predictions[start_idx + idx]
        actual = all_actuals[start_idx + idx]
        error = all_errors[start_idx + idx]
        plt.title(f'Dataset {dataset_idx}\nPred: {pred:.1f}\nTrue: {actual:.1f}\nError: {error:.2f}')
        plt.axis('off')

plt.tight_layout()
plt.show()
```
